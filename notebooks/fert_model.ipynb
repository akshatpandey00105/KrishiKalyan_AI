```python
import pandas as pd
df_fert = pd.read_csv("fertilizer.csv")
```

```python
df_fert.head()
```

```python
df_fert = df_fert.drop(columns=["District_Name", "Link"])
```

```python
df_fert["Crop"] = df_fert["Crop"].str.replace(" ", "")
df_fert["Crop"] = df_fert["Crop"].str.lower().str.strip()
```

```python
mapping = {
    "moong": "mungbean",
    "urad": "blackgram",
    "masoor": "lentil",
    "tur": "pigeonpeas",
    "gram": "chickpea",
    "groundnut": "groundnut",  # already fine after space removal
    "jowar": "millet"
}
df_fert["Crop"] = df_fert["Crop"].replace(mapping)
```

```python
print(df_fert.columns)
```

```python
df_fert = df_fert.rename(columns={
    "Soil_color": "Soil",
    "pH": "PH",
    "Phosphorus": "Phosphorous"
})
```

```python
print(set(df_fert["Crop"].unique()))
```

```python
df_fert = df_fert[df_fert["Crop"].isin([
    'rice','maize','blackgram','cotton','chickpea','sugarcane',
    'lentil','millet','grapes','pigeonpeas','groundnut','mungbean','wheat'
])]
```

```python
df_fert["Soil"] = (
    df_fert["Soil"]
    .str.strip()
    .str.lower()
)

# mapping
soil_mapping = {
    "black": "Alkaline",
    "red": "Acidic",
    "dark brown": "Peaty",
    "medium brown": "Loamy",
    "light brown": "Neutral",
    "reddish brown": "Acidic"
}

# apply mapping safely
df_fert["Soil"] = df_fert["Soil"].map(soil_mapping).fillna(df_fert["Soil"])
```

```python
df_fert["Soil"].unique()
```

```python
df_fert.head()
```

```python
df_fert["Crop"] = df_fert["Crop"].replace("rice", "paddy")
```

```python
    # split
X_fert = df_fert.drop("Fertilizer", axis=1)
y_fert = df_fert["Fertilizer"]

# encode categorical columns
X_fert = pd.get_dummies(X_fert, columns=["Soil", "Crop"])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

X_train, X_test, y_train, y_test = train_test_split(
    X_fert, 
    y_fert, 
    test_size=0.2, 
    random_state=42,
    stratify=y_fert
)
```

```python
rf_fert = RandomForestClassifier(n_estimators=100, random_state=42,class_weight="balanced")
rf_fert.fit(X_train, y_train)
y_pred = rf_fert.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score

# Accuracy (you already trained rf_fert)
y_pred = rf_fert.predict(X_test)

print("Test Accuracy:", accuracy_score(y_test, y_pred))
```
